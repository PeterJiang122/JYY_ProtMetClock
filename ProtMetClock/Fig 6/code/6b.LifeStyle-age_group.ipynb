# Smoking

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
from matplotlib.patches import Patch

# Read CSV file
df = pd.read_csv('../data/AR+Age+Pre_Age+LifeStyle_Score/1.AR+Age+Pre_Age+Smoking_Score.csv')

# Filter data for participants aged 40 and older
df = df[df['Age'] >= 40]

# Create age groups
bins = [40, 50, 60, 70]
labels = ['40-49', '50-59', '60-69']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Create figure with specified dimensions
plt.figure(figsize=(6, 5))

# Create split violin plot showing predicted age distributions
sns.violinplot(
    x='Age Group',
    y='Pre_Age',
    hue='Smoking Score',
    gap=.1,
    data=df,
    palette={0: '#CC8963',
            1: '#5975A4'},
    hue_order=[0, 1],
    split=True,
    inner="quart",
    linewidth=2,
    width=1
)

# Calculate and annotate p-values for each age group
for i, group in enumerate(labels):
    # Split data by smoking status
    group_data = df[df['Age Group'] == group]
    smoking = group_data[group_data['Smoking Score'] == 0]['Pre_Age']
    non_smoking = group_data[group_data['Smoking Score'] == 1]['Pre_Age']
    
    # Perform Welch's t-test
    stat, p_value = ttest_ind(non_smoking, smoking, equal_var=False)
    
    # Determine y-position for annotation
    y_max = max(smoking.max(), non_smoking.max())
    y_pos = y_max + 5
    
    # Format p-value text
    if p_value < 0.001:
        p_text = 'p < 0.001'
    else:
        p_text = f'p = {p_value:.3f}'
    
    # Add annotation to plot
    plt.text(
        x=i,
        y=y_pos,
        s=p_text,
        ha='center',
        fontsize=12
    )

# Add title and axis labels
plt.title('Predicted Age by Smoking Status', fontsize=18, pad=15)
plt.xlabel('Age Group (years)', fontsize=16)
plt.ylabel('Predicted Age (years)', fontsize=16)

# Create custom legend
red_patch = Patch(color='#5975A4', label='Non-Smoking')
blue_patch = Patch(color='#CC8963', label='Smoking')
plt.legend(
    handles=[blue_patch, red_patch],
    fontsize=14,
    loc='upper left',
    frameon=False
)

# Set axis limits and ticks
plt.ylim(27, 98)
plt.yticks(np.arange(30, 91, 10))
plt.tick_params(axis='both', labelsize=14)

# Customize axis borders
ax = plt.gca()
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)

# Remove top and right borders
sns.despine(top=True, right=True)

# Adjust layout and display
plt.tight_layout()
plt.show()
```

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
from matplotlib.patches import Patch

# Read CSV file
df = pd.read_csv('../data/AR+Age+Pre_Age+LifeStyle_Score/9.AR+Age+Pre_Age+Social_Score.csv')

# Filter data to include only participants aged 40 and above
df = df[df['Age'] >= 40]

# Create age group categories
bins = [40, 50, 60, 70]
labels = ['40-49', '50-59', '60-69']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Initialize figure with specified dimensions
plt.figure(figsize=(6, 5))

# Create split violin plot comparing predicted age distributions by social engagement
sns.violinplot(
    x='Age Group',
    y='Pre_Age',
    hue='Social Score',
    gap=.1,
    data=df,
    palette={0: '#CC8963', 1: '#5975A4'},
    hue_order=[0, 1],
    split=True,
    inner="quart",
    linewidth=1
)

# Calculate and annotate statistical significance for each age group
for i, group in enumerate(labels):
    # Subset data for current age group
    group_data = df[df['Age Group'] == group]
    
    # Split into groups based on social engagement
    low_social = group_data[group_data['Social Score'] == 0]['Pre_Age']
    high_social = group_data[group_data['Social Score'] == 1]['Pre_Age']
    
    # Perform Welch's t-test
    stat, p_value = ttest_ind(low_social, high_social, equal_var=False)
    
    # Determine position for p-value annotation
    y_max = max(low_social.max(), high_social.max())
    y_pos = y_max + 5
    
    # Format p-value text according to scientific conventions
    if p_value < 0.001:
        p_text = 'p < 0.001'
    else:
        p_text = f'p = {p_value:.3f}'
    
    # Add annotation to the plot
    plt.text(
        x=i,
        y=y_pos,
        s=p_text,
        ha='center',
        fontsize=12
    )

# Configure plot title and axis labels
plt.title('Biological Age by Social Engagement Level', fontsize=18, pad=15)
plt.xlabel('Age Group (years)', fontsize=15)
plt.ylabel('Predicted Biological Age (years)', fontsize=15)

# Create custom legend with more descriptive labels
red_patch = Patch(color='#CC8963', label='Low Social Engagement')
blue_patch = Patch(color='#5975A4', label='High Social Engagement')
plt.legend(
    handles=[red_patch, blue_patch],
    fontsize=12,
    loc='upper left',
    frameon=False
)

# Set y-axis limits and tick marks
plt.ylim(27, 98)
plt.yticks(np.arange(30, 91, 10))
plt.tick_params(axis='both', labelsize=13)

# Remove top and right borders for cleaner appearance
sns.despine(top=True, right=True)

# Adjust layout to prevent label clipping
plt.tight_layout()

# Display the final visualization
plt.show()
```
