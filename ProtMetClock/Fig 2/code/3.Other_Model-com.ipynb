```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read CSV file containing model comparison data
# Expected columns: Model names and Pearson correlation values
df = pd.read_csv('../data/3.Other_Model-com.csv')

# Prepare data for visualization
models = df['Model']
r = df['Pearson_correlation']  # Extract Pearson correlation values
bar_width = 0.5

# Create figure with specified dimensions (width, height in inches)
fig, ax = plt.subplots(figsize=(5, 4))  # Adjust size as needed

# Create horizontal bar plot
# Using distinct colors from matplotlib's default color cycle
bars = ax.barh(models, r, height=bar_width, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'], alpha=0.7)

# Configure axis labels
ax.set_xlabel('Pearson correlation', fontsize=15)

# Configure axis positions
ax.xaxis.set_ticks_position('bottom')
ax.xaxis.set_label_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.yaxis.set_label_position('left')

# Remove top and right spines for cleaner appearance
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Set x-axis limits to focus on relevant correlation range
ax.set_xlim([0.89, 0.93])

# Customize y-axis tick labels with model abbreviations
ax.set_yticklabels(['SR', 'Organage', 'Paum', 'Milton'], fontsize=13)

# Adjust tick label sizes for better readability
plt.tick_params(axis='x', labelsize=13)
plt.tick_params(axis='y', labelsize=13)

# Automatically adjust subplot parameters to prevent label overlap
plt.tight_layout()

# Display the plot
plt.show()
```
